```python
#Leitura da base de dados
import pandas as pd

df = pd.read_csv('/Workspace/Users/glauber.bomtempo@samarco.com/7DaysOfCode/dataset.csv')  
df.head()
```

```python
#ver informações gerias do dataframe
df.info()
```

```python
#excluir linhas com valores nulos
df = df.dropna(subset=['artists', 'album_name', 'track_name'])
```

```python
#ver informações gerias do dataframe após as exclusões
df.info()
```

```python
#excluir colunas desnecessárias
if 'Unnamed: 0' in df.columns:    df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
#verificando os valores nulos por coluna
df.isnull().sum().sort_values(ascending=False)
```

```python
#verificando se há linhas duplicadas
df.duplicated().sum()
```

```python
#removendo duplicados
df = df.drop_duplicates()
df.duplicated().sum()  # deve retornar 0
```

```python
#estatísticas descritivas
df.describe().T
```

```python
# Remover músicas com duração maior que 30 minutos (1.800.000 ms)
# Aceitamos músicas com até 30 minutos (como as longas do Pink Floyd ou Iron Maiden)
df = df[df['duration_ms'] <= 1800000]

# Remover músicas com tempo (BPM) inválido ou ausente
# Valores válidos são maiores que 0 BPM (música sem batida identificável é inválida)
df = df[df['tempo'] > 0]

# Remover músicas com time signature inválido
# Valores válidos são 3, 4, 5... (não existe compasso 0)
df = df[df['time_signature'] > 0]

# Remover músicas com loudness (volume médio) absurdamente baixo
# Valores normais variam entre -5 dB (alto) até -30 dB (baixo); usamos -40 dB como limite
df = df[df['loudness'] > -40]
```

```python
#ver informações gerias do dataframe após as exclusões
df.info()
```

```python
#estatísticas descritivas após as exclusões
df.describe().T
```

```python
#Criando coluna binária de popularidade com um corte de popularidade 50. Pelas estatísticas descritivas, 50 é o valor que divide o 4ªquartil do 3ºquartil. Isso tornará a base de dados um pouco desbalanceada, mas retrata melhor a realidade, onde temos muito mais músicas com popularidades médias ou baixas do que Hits de sucesso. Para lidar bem com esse desbalanceamento na base de dados pretendo utilizar modelos de classificação como florestas aleatórias ou XGBoost que lidam bem com dados debalanceados.


import numpy as np

# Definir o corte em 50 para considerar 'popular'
corte_popularidade = 50

# Criar coluna binária: 1 = popular (>=50), 0 = não popular (<50)
df['popular_binaria'] = np.where(df['popularity'] >= corte_popularidade, 1, 0)

# Visualizar contagem
df['popular_binaria'].value_counts(normalize=True)
```

```python
#Remoção de colunas que não serão utilizadas no modelo

# Lista inicial de colunas para remover
colunas_remover = ['track_id', 'track_name', 'album_name', 'popularity']

# Remover colunas
df_modelo = df.drop(columns=colunas_remover)

# Conferir dimensões após exclusão
print(f'Dataset após remoção: {df_modelo.shape}')
df_modelo.head()
```

```python
#Essa etapa calcula a média de popularidade por artista e seleciona os 5000 artistas mais populares. Em seguida, cria uma coluna binária que indica se um artista é relevante ou não se a música tiver participação de um desses 5000 artistas. Depois removo a coluna inicial artists.

# Passo 1: separar os artistas
df['artists_list'] = df['artists'].str.split(';')

# Explodir linhas para calcular média de popularidade por artista
df_explodido = df.explode('artists_list')

# Calcular média de popularidade por artista
media_artistas = (
    df_explodido.groupby('artists_list')['popularity']
    .mean()
    .sort_values(ascending=False)
)

# Selecionar top 5000 artistas mais populares
top_5000_artistas = set(media_artistas.head(5000).index)

# Passo 2: definir função para criar coluna binária
def marcar_artista_relevante(lista_artistas):
    if any(artista.strip() in top_5000_artistas for artista in lista_artistas):
        return 1
    else:
        return 0

# Passo 3: criar coluna binária
df_modelo['artistas_relevantes'] = df['artists_list'].apply(marcar_artista_relevante)

# Remover coluna original 'artists' e 'artists_list'
df_modelo = df_modelo.drop(columns=['artists'])

# Conferir resultado
df_modelo['artistas_relevantes'].value_counts(normalize=True)
```

```python
# OneHotEncoding para track_genre
df_modelo = pd.get_dummies(df_modelo, columns=['track_genre'], drop_first=True)

print(f'Dataset após OneHotEncoding: {df_modelo.shape}')
df_modelo.head()
```

```python
#Modelo de Florestas Aleatórias - Apliquei o RandomOverSample para fazer uma reamostragem dos dados para melhor balanceá-los e depois utilzei esse modelo com a divisão de dados com validação cruzada 

# Imports necessários
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import RandomOverSampler
import numpy as np

# Separando treino e teste (20% para teste)
X = df_modelo.drop(columns=['popular_binaria'])
y = df_modelo['popular_binaria']

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(f"Tamanho treino: {X_treino.shape}, Tamanho teste: {X_teste.shape}")
print(f"Distribuição treino:\n{y_treino.value_counts(normalize=True)}")
print(f"Distribuição teste:\n{y_teste.value_counts(normalize=True)}")

# Aplicando RandomOverSampler no conjunto de treino
ros = RandomOverSampler(random_state=42)
X_treino_res, y_treino_res = ros.fit_resample(X_treino, y_treino)

print(f"\nApós RandomOverSampler - Treino balanceado:")
print(y_treino_res.value_counts(normalize=True))

# Definindo StratifiedKFold para validação cruzada balanceada
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Treinando RandomForest com validação cruzada usando dados balanceados
accuracies, precisions, recalls, f1_scores = [], [], [], []

for fold, (train_idx, val_idx) in enumerate(skf.split(X_treino_res, y_treino_res)):
    X_fold_train, X_fold_val = X_treino_res.iloc[train_idx], X_treino_res.iloc[val_idx]
    y_fold_train, y_fold_val = y_treino_res.iloc[train_idx], y_treino_res.iloc[val_idx]
    
    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    model.fit(X_fold_train, y_fold_train)
    y_pred = model.predict(X_fold_val)

    acc = accuracy_score(y_fold_val, y_pred)
    prec = precision_score(y_fold_val, y_pred)
    rec = recall_score(y_fold_val, y_pred)
    f1 = f1_score(y_fold_val, y_pred)

    accuracies.append(acc)
    precisions.append(prec)
    recalls.append(rec)
    f1_scores.append(f1)

    print(f"\nFold {fold+1}")
    print(f"Acurácia: {acc:.4f}, Precisão: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

print("\n✅ Resultados Médios após Oversampling com RandomForest:")
print(f"Acurácia média: {np.mean(accuracies):.4f}")
print(f"Precisão média: {np.mean(precisions):.4f}")
print(f"Recall médio: {np.mean(recalls):.4f}")
print(f"F1-score médio: {np.mean(f1_scores):.4f}")

```

```python
# Prevendo no conjunto de teste
y_pred_teste = model.predict(X_teste)

# Avaliando desempenho no teste
acc_teste = accuracy_score(y_teste, y_pred_teste)
prec_teste = precision_score(y_teste, y_pred_teste)
rec_teste = recall_score(y_teste, y_pred_teste)
f1_teste = f1_score(y_teste, y_pred_teste)

print("✅ Resultados no conjunto de teste (20% não vistos):")
print(f"Acurácia: {acc_teste:.4f}")
print(f"Precisão: {prec_teste:.4f}")
print(f"Recall: {rec_teste:.4f}")
print(f"F1-score: {f1_teste:.4f}")
```

## 📝 Procedimento de Preparação da Nova Observação para Predição

Para garantir que a nova observação (exemplo: uma música recém-lançada) seja corretamente processada pelo modelo treinado, é necessário realizar uma série de transformações de dados **antes da predição**, respeitando as mesmas etapas aplicadas ao conjunto de dados utilizado durante o treinamento.

---

### ✅ Etapas do Pré-Processamento da Nova Observação

#### 1. Leitura da Nova Observação:
A nova observação será fornecida através de uma **planilha Excel ou CSV** contendo **uma única linha** e inicialmente **19 colunas**, incluindo informações como:

- `track_id`, `track_name`, `album_name`, `artists`, `track_genre`, além de colunas numéricas como `danceability`, `energy`, `loudness`, etc.

---

#### 2. Exclusão de Colunas Irrelevantes:
As colunas **`track_id`**, **`track_name`** e **`album_name`** serão **removidas**, pois não contribuem para o modelo preditivo e servem apenas para identificação da faixa.

---

#### 3. Criação da Coluna `artistas_relevantes`:
- A coluna `artists` será comparada a uma **lista fixa** dos **5000 artistas mais populares** extraída da base de dados usada no treinamento.

- **Regra aplicada**:
    - Se o artista estiver presente na lista dos 5000 artistas mais populares → `artistas_relevantes = 1`
    - Caso contrário → `artistas_relevantes = 0`

- Após a criação desta feature, a coluna original `artists` será **descartada**.

---

#### 4. Conversão da Coluna `track_genre` via One-Hot Encoding:
- A coluna `track_genre` deverá conter um **único gênero musical** selecionado por lista suspensa com **113 opções pré-definidas** (mesma lista usada no treinamento).
- Esta coluna será convertida em **113 colunas binárias (one-hot encoding)**, preenchendo **1** na coluna correspondente ao gênero informado e **0** nas demais 112 colunas.
- A coluna original `track_genre` será **removida** após a conversão.

---

#### 5. Formatação Final da Nova Observação:
A planilha finalizada deverá conter exatamente **128 colunas (features)**, mantendo consistência com o dataset original usado no treinamento do modelo.

**Exemplo da estrutura final**:
- 19 colunas originais
- **– 3 colunas excluídas** (`track_id`, `track_name`, `album_name`)
- **– 1 coluna excluída** (`artists`)
- **+ 1 coluna adicionada** (`artistas_relevantes`)
- **– 1 coluna excluída** (`track_genre`)
- **+ 113 colunas binárias** (one-hot encoding do `track_genre`)

✅ **Total final = 19 - 3 - 1 + 1 - 1 + 113 = 128 features**

---

#### 6. Ordem e Nome das Colunas:
- O **ordenamento das colunas** na planilha resultante deverá seguir a **mesma ordem exata** das colunas usadas no treinamento do modelo para evitar erros na predição.
- O pipeline de predição deve incluir uma **validação/verificação automática** da ordem correta das colunas antes de alimentar o modelo.

---

### 🎯 Observações Importantes
- Este pré-processamento é **fundamental** para garantir a **coerência dos dados de entrada** com o modelo treinado.
- Recomendável implementar este pipeline em **Python** como uma função ou classe, evitando erros manuais.
- O pipeline pode ser facilmente adaptado para fluxos de produção (Dash, Streamlit, Gradio ou APIs Flask/FastAPI).
